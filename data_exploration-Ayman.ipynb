```python
#plot a map of the data
import folium
from folium.plugins import HeatMap

# Make an empty map
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], tiles="CartoDB dark_matter", zoom_start=11)

#plot dots on the map by latitude and longitude
def plot_dots(latm, longm):
    folium.Circle(
        location=[latm, longm],
        radius=1,
        color='crimson',
        fill=True,
        fill_color='crimson'
    ).add_to(m)

# I can add marker one by one on the map
def plot_df_dots(df):
    for i in range(0,len(df)):
        plot_dots(df.iloc[i]['latitude'], df.iloc[i]['longitude'])


plot_df_dots(df)

#plot lines between points

def plot_lines(lat1, long1, lat2, long2):
    folium.PolyLine(locations=[[lat1, long1], [lat2, long2]], color="red", weight=2.5, opacity=0.2).add_to(m)

def plot_df_lines(df):
    for i in range(0,len(df)-1):
        plot_lines(df.iloc[i]['latitude'], df.iloc[i]['longitude'], df.iloc[i+1]['latitude'], df.iloc[i+1]['longitude'])

#plot_df_lines(df)

#plot a heatmap of the data
# List comprehension to make out list of lists
# Plot it on the map
def plotheat(df):
    heat_data = [[row['latitude'],row['longitude']] for index, row in df.iterrows()]
    HeatMap(heat_data).add_to(m)

plotheat(df)


# Save it as html
m.save('mymap.html')

```
